#### 2. Convert Columns to correct Data types ( string to numeric here) 

```python
#Convert Columns to correct data type

cust['Customer_Age']= pd.to_numeric(cust['Customer_Age']) 


cust['Income']= pd.to_numeric(cust['Income'])


cust['Dependent_Count']= pd.to_numeric(cust['Dependent_Count'])


```

```python
#Convert Columns  of Cc table to correct data type

cc['Total_Trans_Amt']= pd.to_numeric(cc['Total_Trans_Amt']) 

cc['Interest_Earned']= pd.to_numeric(cc['Interest_Earned'])

cc['Annual_Fees']= pd.to_numeric(cc['Annual_Fees'])

cc.head()
```

#### 3. creating new column Age group based on following condition using "apply"

```python

def val (d):
    if d['Customer_Age'] < 30:
        return "<30"
    elif d['Customer_Age'] >=30 and d['Customer_Age']<40:
        return "30-40"
    elif d['Customer_Age']>=40 and d['Customer_Age']<50:
        return "40-50"
    elif d['Customer_Age'] >=5 and d['Customer_Age']<60:
        return "50-60"
    elif d['Customer_Age'] >=60:
        return ">=60"
    
cust['Age_Group'] = cust.apply(val, axis =1)
    
    

cust.head() 
```

```python
#creating new column ' Income Group ' based on following condition using "apply"


def val (d):
    if d['Income'] < 35000:
        return "Low"
    elif d['Income'] >=35000 and d['Income']<70000:
        return "Medium"
    elif d['Income']>=70000:
        return "High"
    
           
           
cust['Income_Group'] = cust.apply(val, axis =1)
    
    
cust.head() 
```

```python
cc.head()
```

#### 4.creating new column "Total Revenue" I. e.  Total Revenue generated by bank for its credit card services to the customers. 

####Total Revenue = Anuual Fees + Interest earned + Total Transaction Amount.

```python


cc['Total_Revenue']= cc['Annual_Fees'] + cc['Total_Trans_Amt'] + cc['Interest_Earned']

cc.head()
```

#### 3. Activation 30 days with card category & Gender



# Merge two data frames cust & cc 

```python
import pandas as pd
import matplotlib.pyplot as plt
# Assuming you have the following DataFrames:
# cust = pd.read_csv('customer_details.csv')  # Customer details DataFrame
# cc = pd.read_csv('credit_card_details.csv')  # Credit card details DataFrame
# Merge the two DataFrames on Client ID

merged = pd.merge(cc, cust[['Client_Num', 'Gender']], on='Client_Num')


```

```python
activation_rate = merged.groupby(['Card_Category', 'Gender'])['Activation_30_Days'].mean().unstack()
activation_rate.plot(kind='bar', color=['Red', 'darkblue'])
plt.title('Activation Rate within 30 Days by Category and Gender')
plt.xlabel('Category')
plt.ylabel('Activation Rate')
plt.show()
```
